```python
import os, sys
import torch
from PIL import Image
import matplotlib.pyplot as plt
sys.path.append("..")
from subject_driven_inpainting import generate_image, get_flux_pipe, load_loras, save_image
```

```python
# set up parameters
pipe = get_flux_pipe("cpu")
pipe.to("cuda:0")
lora_path1 = ""# path to background inpainting lora (fill)
lora_path2 = ""# path to subject lora (subject)

# load loras
load_loras(pipe, lora_path1, lora_path2)

```

```python
path_background = "./bg.png"
path_subject = "./subject.jpg"

# generate image
background = Image.open(path_background)
background = background.resize((512, 512))
os.makedirs("./results", exist_ok=True)
background.save("./results/background.png")
# Display the background image
plt.figure(figsize=(10, 10))
plt.imshow(background)
plt.axis('off')
plt.title("Background Image")
# Draw bounding box
# Define bbox
bbox = [165, 160, 265, 355]  # [x, y, x+width, y+height] based on Rectangle(x=160, y=170, width=110, height=185)
rect = plt.Rectangle((165, 160), 100, 195, fill=False, color='red', linewidth=2)
plt.gca().add_patch(rect)
plt.show()
```

```python
prompt = "Fill the blank area in the given background with A Chinese traditional dress., with description: Amidst the bright sunshine."
bg_path = ""
subject_path = ""
save_path = "./results/generated_result.png"

torch.manual_seed(421)

# Generate and save image
result_image = generate_image(prompt, bg_path, subject_path, bbox, pipe)
save_image(result_image, save_path)

# Display the result
plt.figure(figsize=(10, 10))
plt.imshow(result_image)
```
